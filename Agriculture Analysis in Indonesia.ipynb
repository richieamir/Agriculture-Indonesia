### Importing Packages

```python
import tabula
import pandas as pd
import os
```

## Read production data 2018

```python
production_2018 = os.listdir('data/production/2018') 
production_2019 = os.listdir('data/production/2019')
```

```python
dfs = {}
for plant in production_2018:
    temp = plant.split('.')
    var_name = 'df-' + temp[0]
    
    df = tabula.read_pdf("data/production/2018/{}".format(plant), pages='all')
    df = pd.DataFrame(df[0])
    df = df.drop([0,1,2,3,4])
    df = df.rename(columns={'Unnamed: 1':'Province','Unnamed: 4':'Production','Unnamed: 6':'Production_growth'})
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df = df.drop(['Tahun'],axis=1)
    df = df.dropna()
    df.reset_index(inplace=True,drop=True)
    dfs[var_name] = df
```

## Read Production data 2019

```python
for plant in production_2019:
    temp = plant.split('.')
    var_name = 'df-' + temp[0]
    
    df = tabula.read_pdf("data/production/2019/{}".format(plant), pages='all')
    df = pd.DataFrame(df[0])
    df = df.drop([0,1,2,3,4])
    df = df.rename(columns={'Unnamed: 1':'Province','Unnamed: 6':'Production','Pertumbuhan/':'Production_growth'})
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
#     df = df.drop(['Tahun'],axis=1)
    df = df.dropna()
    df.reset_index(inplace=True,drop=True)
    dfs[var_name] = df
```

```python
for key in dfs:
   print("key: {}".format(key))
```
