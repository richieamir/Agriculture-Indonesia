### Importing Packages

```python
import tabula
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read production data 2018

```python
production_2018 = os.listdir('data/production/2018') 
production_2019 = os.listdir('data/production/2019')
```

```python
df_agriculture = []
df_agriculture = pd.DataFrame(df_agriculture)
```

```python
dfs = {}
counter = 0;
for plant in production_2018:
    temp = plant.split('.')
    var_name =  temp[0].replace("-production","")
    
    df = tabula.read_pdf("data/production/2018/{}".format(plant), pages='all')
    df = pd.DataFrame(df[0])
    df = df.drop([0,1,2,3,4])
    if counter == 0:
        df = df.rename(columns={'Unnamed: 1':'Province','Unnamed: 4':'{}'.format(var_name)})
    else:
        df = df.rename(columns={'Unnamed: 4':'{}'.format(var_name)})
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df = df.drop(['Tahun'],axis=1)
    df = df.dropna()
    df.reset_index(inplace=True,drop=True)
    if counter == 0:
        df_agriculture = df
    else:
        df_agriculture[var_name] = df[var_name]
    counter += 1
#     dfs[var_name] = df
#     df_agriculture[var_name] = df
```

## Read Production data 2019

```python
for plant in production_2019:
    temp = plant.split('.')
    var_name =  temp[0].replace("-production","")
    
    df = tabula.read_pdf("data/production/2019/{}".format(plant), pages='all')
    df = pd.DataFrame(df[0])
    df = df.drop([0,1,2,3,4])
    if var_name == 'chrysanthemum' or var_name == 'garlic':
        df = df.rename(columns={'Unnamed: 8':'{}'.format(var_name)})
    else:
        df = df.rename(columns={'Unnamed: 6':'{}'.format(var_name)})
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
#     df = df.drop(['Tahun'],axis=1)
    df = df.dropna()
    df.reset_index(inplace=True,drop=True)
    df_agriculture[var_name] = df[var_name]
```
